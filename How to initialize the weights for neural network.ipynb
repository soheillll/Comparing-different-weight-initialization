# How Initialize Weights for Our Neural Network

In here we will learn how to initialize our neural network's weight in a way that it will boost the performance of our model. In this lesson we will initialize the network with the following options:
    1. Initializing all with 0
    2. Initializing all with 1
    3. Initializing with uniform distribution
    4. Initializing with normal distribution
    5. Initializing with truncated normal distribution
    
Before going through all of mensioned steps, let's write our initial codes.

```python
# Importig the libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from helper_func import get_the_result, _get_loss_acc
from tensorflow.examples.tutorials.mnist import input_data

%matplotlib inline
```

```python
# Loading the MNIST dataset
mnist = input_data.read_data_sets("MNIST_data", one_hot = True)
```

## 1. Initializing all with 0

```python
all_zero_weights = [tf.Variable(tf.zeros((mnist.train.images.shape[1], 512))),
                    tf.Variable(tf.zeros((512, 256))),
                    tf.Variable(tf.zeros((256, mnist.train.labels.shape[1])))]

get_the_result(dataset = mnist, 
               title="Initializing weights with 0", 
               weight_init_list = [(all_zero_weights, "Zeros Weights")], 
               plot_n_batches = 100)
```

## 2. Initializing all with 1

```python
all_ones_weights = [tf.Variable(tf.ones((mnist.train.images.shape[1], 512))),
                    tf.Variable(tf.ones((512, 256))),
                    tf.Variable(tf.ones((256, mnist.train.labels.shape[1])))]

get_the_result(dataset = mnist, 
               title="Initializing weights with 1", 
               weight_init_list = [(all_ones_weights, "Ones Weights")], 
               plot_n_batches = 100)
```

## 3. Initializing with uniform distribution

```python
uniform_distribution_weights = [tf.Variable(tf.random_uniform((mnist.train.images.shape[1], 512), -1, 1)),
                    tf.Variable(tf.random_uniform((512, 256), -1, 1)),
                    tf.Variable(tf.random_uniform((256, mnist.train.labels.shape[1]), -1, 1))]

get_the_result(dataset = mnist, 
               title="Initializing weights with uniform distribution", 
               weight_init_list = [(uniform_distribution_weights, "uniform distribution Weights")], 
               plot_n_batches = 100)
```

## 4. Initializing with normal distribution

```python
normal_distribution_weights = [tf.Variable(tf.random_normal((mnist.train.images.shape[1], 512), stddev = 0.1)),
                    tf.Variable(tf.random_normal((512, 256), stddev = 0.1)),
                    tf.Variable(tf.random_normal((256, mnist.train.labels.shape[1]),stddev = 0.1))]

get_the_result(dataset = mnist, 
               title="Initializing weights with normal distribution", 
               weight_init_list = [(normal_distribution_weights, "normal distribution Weights")], 
               plot_n_batches = 100)
```

## 5. Initializing with truncated normal distribution

```python
truncated_normal_distribution_weights = [tf.Variable(tf.truncated_normal((mnist.train.images.shape[1], 512), stddev = 0.1)),
                    tf.Variable(tf.truncated_normal((512, 256), stddev = 0.1)),
                    tf.Variable(tf.truncated_normal((256, mnist.train.labels.shape[1]),stddev = 0.1))]

get_the_result(dataset = mnist, 
               title="Initializing weights with truncated normal distribution", 
               weight_init_list = [(truncated_normal_distribution_weights, "truncated normal distribution Weights")], 
               plot_n_batches = 100)
```

## All in one graph

```python
truncated_normal_distribution_weights = [tf.Variable(tf.truncated_normal((mnist.train.images.shape[1], 512), stddev = 0.1)),
                    tf.Variable(tf.truncated_normal((512, 256), stddev = 0.1)),
                    tf.Variable(tf.truncated_normal((256, mnist.train.labels.shape[1]),stddev = 0.1))]

get_the_result(dataset = mnist, 
               title="Initializing weights with truncated normal distribution", 
               weight_init_list = [(truncated_normal_distribution_weights, "truncated normal distribution Weights"),
                                   (normal_distribution_weights, "normal distribution Weights"),
                                   (uniform_distribution_weights, "uniform distribution Weights"),
                                   (all_ones_weights, "Ones Weights"),
                                   (all_zero_weights, "Zeros Weights")], 
               plot_n_batches = 100)
```
